```python
from sklearn.preprocessing import StandardScaler
```

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.linear_model import LinearRegression
```

```python
from sklearn.pipeline import make_pipeline
```

```python
from sklearn.compose import make_column_transformer
```

```python
lr =LinearRegression()
```

```python
scaler = StandardScaler()
```

```python
model = make_pipeline(scaler, lr)
```

```python
data_input = data.drop( columns = ['SalePrice'])
data_output = data['SalePrice']
```

```python
x_train,x_test,y_train,y_test = train_test_split(data_input, data_output,test_size = 0.2)
```

```python
model.fit(x_train, y_train)
```

```python
model.score(x_test,y_test)
```

```python
input = pd.DataFrame([[1300,8550,3,3]], columns = ['Id','LotArea','FullBath','BedroomAbvGr'])
```

```python
model.predict(input)
```
